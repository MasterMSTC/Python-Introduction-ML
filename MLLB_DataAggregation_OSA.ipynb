## Read Excel File with Clinical Variables for exteme male cases

```python
# Import pandas
import pandas as pd

# Read Excell with extreme male cases
file = 'OSA_extreme_male.xlsx'

# Load spreadsheet
xl = pd.ExcelFile(file)

# Print the sheet names
print(xl.sheet_names)
```

```python

# Load a sheet into a DataFrame by name: df_OSA_extreme
df_OSA_extreme = xl.parse('Sheet1')
```

```python
df_OSA_extreme.head(5)
```

## Read Excel File with Speech Variables for male cases

```python
# Read Excell with extreme male cases
file = 'Formants_Bandwidth_OSA_male_speakers.xlsx'

# Load spreadsheet
xl = pd.ExcelFile(file)

# Print the sheet names
print(xl.sheet_names)
```

```python

# Load a sheet into a DataFrame by name: df_OSA_speech
df_OSA_speech = xl.parse('SPEECH_FEATURES')
```

```python
df_OSA_speech.head(5)
```

```python
df_OSA_speech.columns
```

## Compare the two Data Frames

```python
df_OSA_extreme.describe()
```

```python
df_OSA_speech.describe()
```

## Give the same name to the common column (index):
   **Patient**

```python
df_OSA_speech = \
    df_OSA_speech.rename(columns = {"PATIENT": "Patient"})
```

```python
df_OSA_speech['Patient'].describe()
```

```python
df_OSA_extreme['Patient'].describe()
```

```python
df_OSA_extreme.info()
```

```python
df_OSA_speech.info()
```

## Learn how to join data frames
* ### See: MLLB_DataAgregation_Join_DataFrames.pptx

```python
# Make an inner join ON 'Patient'
df_OSA_inner = pd.merge(df_OSA_extreme, df_OSA_speech, on='Patient', how='inner')
```

```python
df_OSA_inner.info()
```

## Learn how to use groupby in pandas data frames
* ### See, for example [Summarising, aggregation and grouping data in python pandas](https://www.shanelynn.ie/summarising-aggregation-and-grouping-data-in-python-pandas/)

```python
df_OSA_inner.groupby('OSA')['OSA'].describe()
```

```python
df_OSA_inner.groupby('OSA')['BMI'].describe()
```

## Write df_OSA_inner to an Excel file

```python
df_OSA_inner.to_excel (r'OSA_Aggregated_UPM.xlsx', index = None, header=True) 
```
